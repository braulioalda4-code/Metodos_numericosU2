```python
#PARTE A
import numpy as np #Biblioteca para cálculo numérico
import matplotlib.pyplot as plt

def f(x): #Definicion de la función matemática
    return x**3 - 4 * x - 9 # Expresión algebraica de la función

a = 2 # Límite inferior del intervalo
b = 3 # Límite superior del intervalo

fa = f(a) # Evaluación de la función en el extremo izquierdo
fb = f(b) # Evaluación de la función en el extremo derecho

print("f(a):", fa) #Mostrar valor en a
print("f(b):", fb)

if fa * fb < 0: # Verifica condición del Teorema del valor intermedio
  print("Exite al menos una raíz en el intervalo") # Confirmación
else:
  print("No se garantiza raíz en el intervalo") # No hay cambio de signo

x = np.linspace(a, b, 100)
y = x**3 - 4 * x - 9

plt.axhline(0)
plt.plot(x, y)
plt.title("Función f(x) = x^3 - x - 2")
plt.show()
```

```python
#PARTE B
import numpy as np

def f(X): # Definición de la función a analizar
    return X**3-4*X-9 # Función no lineal

a = 2 # Extremo izquierdo inicial
b = 3 # Extremo derecho inicial
tolerancia = 1e-5 # Error máximo permitido
max_iter = 100 # Número máximo de iteraciones permitidas

for i in range(max_iter):
    c = (a+b)/2 # Cálculo del punto medio
    fc = f(c) # Evaluación de la función en el punto medio

    if abs(fc) < tolerancia: # Criterio de paro por valor cercano a cero
      break # Termina el ciclo

    if f(a) * fc < 0: # Verifica si la raíz está en el
      b = c # Actualiza extremo derecho
    else: # En caso contrario
        a = c # Actualiza extremo izquierdo

print("Raíz aproximada:", c) # Muestra la raíz encontrada
print("Iteraciones realizadas:", i+1) # Número de iteraciones

a = 2 # Limite inicial inferior
b = 3 # Limite inicial superior
n = 100 # Número de iteraciones

error_teorico = (b-a)/(2**n) # Fórmula del error máximo en bisección

print("Error máximo después de", n, "iteraciones:", error_teorico) # Mostrar error
```

```python
# Parte C
def g(X): # Definición de la función a analizar
  return (4*X+9)**(1/3) # Función no lineal
tolerancia = 1e-5 # Error máximo permitido
X = 2.5 # Valor inicial

for i in range(max_iter): # Rango maximo de iteraciones
  X_nuevo = g(X)

  if abs(X_nuevo - X) < tolerancia: # Criterio de paro por valor cercano a cero
     break # Termina el ciclo

  X = X_nuevo # Valor inicial nuevo

print("Raiz aproximada:", X_nuevo) # Mostrar raiz aproximada
print("Iteraciones realizadas:", i+1) # Mostrar iteraciones realizadas
print("")
print("")
```
