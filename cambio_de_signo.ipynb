Ejercicio 1 - Verificar cambio de signo


```python
import numpy as np #Biblioteca para cálculo numérico

def f(x): #Definicion de la función matemática
    return x**3 - x - 2 # Expresión algebraica de la función

a = 1 # Límite inferior del intervalo
b = 2 # Límite superior del intervalo

fa = f(a) # Evaluación de la función en el extremo izquierdo
fb = f(b) # Evaluación de la función en el extremo derecho

print("f(a):", fa) #Mostrar valor en a
print("f(b):", fb)

if fa * fb < 0: # Verifica condición del Teorema del valor intermedio
  print("Exite al menos una raíz en el intervalo") # Confirmación
else:
  print("No se garantiza raíz en el intervalo") # No hay cambio de signo
```
